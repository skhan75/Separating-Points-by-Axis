```python
import matplotlib.pyplot as plt
from math import log
from operator import itemgetter
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
```

```python
# Reading Text File

files = glob.glob('input/*.txt')
print(files)
for f in files:
    with open(f) as fl:
        text = fl.read().split('\n')
        
# Reading points from text file

n = text[0] # Number of points in the file
points = []
for t in text[1:]:
    x = int(t.split(' ')[0])
    y = int(t.split(' ')[1])
    points.append((x,y))
print(points)
```

```python
points.sort(key=itemgetter(0))
print(points)
V = [13.5,19.5,6.5,15.5,21.5,8.5,1.5,4.5,16.5,25.5]
H = [13.5,18.5,6.5, 23.5]
plt.scatter(*zip(*points))
for v in V:
    plt.axvline(v, color='r')
for h in H:
    plt.axhline(h,color='g')
#plt.ylim([0,1])
plt.show()
```

```python
#Creating graph of all the points
graph = {}
for p in points:
    for j in range(0, len(points)):
        if(points[j] != p):
            #graph[p].append(points[j])
            #graph.setdefault(p, {})[value] = points[j]
            graph.setdefault(p, []).append(points[j])   
```

```python
#Generating Edges to connect points in Graph
edges = []
for node in graph:
    for neighbour in graph[node]:
        edges.append((node, neighbour))
```
